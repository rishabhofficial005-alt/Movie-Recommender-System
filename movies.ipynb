```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import re
```

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
```

```python
!pip install wordninja
import wordninja

def preprocess_text(text):
    text = str(text).lower()

    # keep only letters (remove punctuation + numbers)
    text = re.sub(r'[^a-z]', '', text)

    # 🔥 STEP 1: FIX GLOVED WORDS FIRST
    text = ' '.join(wordninja.split(text))

    # STEP 2: tokenize
    words = text.split()

    # STEP 3: stopwords + lemmatization
    words = [
        lemmatizer.lemmatize(word)
        for word in words
        if word not in stop_words
    ]

    return ' '.join(words)
```

```python
df['tags']=df['tags'].apply(preprocess_text)
```

```python
df['tags'][0]
```

```python
df=df.reset_index(drop=True)
```

```python
indices=pd.Series(df.index,index=df['title']).drop_duplicates()
```

```python
indices
```

```python
###Vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfid_matrix=tfidf.fit_transform(df['tags'])
```

```python
tfid_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title,n=10):
  if title not in indices:
    return ['Movie not found']
  indx=indices[title]
  sim_score=cosine_similarity(tfid_matrix[indx],tfid_matrix).flatten()
  similar_idx=sim_score.argsort()[::-1][1:n+1]
  return df['title'].iloc[similar_idx]
```

```python
recommend('Toy Story')
```

```python
import pickle
pickle.dump(tfid_matrix,open('tfid_matrix.pkl','wb'))
pickle.dump(indices,open('indices.pkl','wb'))
df.to_pickle('df.pkl')
pickle.dump(tfidf,open('tfidf.pkl','wb'))
```
